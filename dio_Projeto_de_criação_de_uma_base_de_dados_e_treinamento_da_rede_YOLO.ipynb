```python
import torch
import matplotlib.pyplot as plt
import cv2

# Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Images
img_path = 'dog_bike.jpg'
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Inference
results = model(img)

# Results
results.print()
results.show()  # or .show()
```
